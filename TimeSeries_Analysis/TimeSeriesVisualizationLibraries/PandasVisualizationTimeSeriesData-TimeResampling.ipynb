## Time Resampling

```python
# Time Resampling Yearly Example
df.resample(rule='YE').mean() # Annual frequency - see documentation for more rules
```

```python
# Time Resampling quarterly Example
df.resample(rule='QE').mean() # Quarterly frequency - see documentation for more rules
```

```python
# Time resampling business quarter end
df.resample(rule='BQE').mean() # Business quarter end frequency - see documentation for more rules
```

```python
# Maximum value resampling yearly
df.resample(rule='YE').max() # Annual frequency - see documentation for more rules
```

```python
# function to return the first entry of each year
def first_entry(entry):
    return entry[0]
```

```python
df.resample('YE').apply(first_entry)
```

```python
df['Close'].resample('YE').mean().plot(kind='bar', figsize=(10,6))
plt.show()
```

```python
df['Close'].resample('ME').mean().plot(kind='bar', figsize=(16,6))
plt.show()
```
