```python
# time series data set
df1 = pd.read_csv('df1',index_col=0)
```

```python
df1.head()
```

```python
# non time series data set
df2 = pd.read_csv('df2')
df2.head()
```

```python
# histogram of a column values for df1
df1['A'].hist(bins=30)
plt.show()
```

```python
df1['A'].plot(kind='hist',bins=30)
plt.show()
```

```python
df1['A'].plot.hist()
plt.show()
```

```python
# area plot for df2
df2.plot.area(alpha=0.4)
plt.show()
```

```python
# bar plot for df2
df2.plot.bar()
plt.show()
```

```python
# stacked bar plot for df2
df2.plot.bar(stacked=True)
plt.show()
```

```python
# line plot for df1 with x and y specified
# either reset index to make it a column again
df1 = df1.reset_index() 
df1.plot.line(x='index', y='B', figsize=(12,3), lw=1)
plt.show()
```

```python
df1.info()
```

```python
# scatter plot for df1 - column A vs B with colour based on C values
df1.plot.scatter(x='A',y='B', c='C', cmap='coolwarm')
plt.show()
```

```python
# bubble plot for df1 - column A vs B with size based on how large C values are, relative to each other
df1.plot.scatter(x='A', y='B', s=df1['C']*100)
plt.show()
```

```python
# boxplot for df2
df2.plot.box()
plt.show()
```

```python
# create a dataframe with random values for a hexagonal bit plot
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
```

```python
df.head()
```

```python
# hexagonal bit plot for df1 for byvariate (2 variable) data with random data
# hexagons gets darker as the number of points in that bin increases
df.plot.hexbin(x='a', y='b', gridsize=25, cmap='coolwarm')
plt.show()
```

```python
# Kernal density estimate plot
df2['a'].plot.kde()
plt.show()
```

```python
# density plot for df2
df2.plot.density()
plt.show()
```

```python
# density plot for df2 for columns a
df2['a'].plot.density(legend=True)
plt.show()
```
