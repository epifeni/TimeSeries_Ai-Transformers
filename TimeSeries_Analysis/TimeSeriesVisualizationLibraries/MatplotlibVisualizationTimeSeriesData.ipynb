## Specify date time index information

```python
fig,ax = plt.subplots()
ax.plot_date(idx, stock, '-')

# fix the date format on x-axis
fig.autofmt_xdate()

# show grid lines
#ax.yaxis.grid(True)
#ax.xaxis.grid(True)

# edit the date format on x-axis - Major ticks
# Locating
ax.xaxis.set_major_locator(dates.MonthLocator()) # set major ticks to monthly interval
# Formatting
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n\n%b %Y')) # set date format to month and year

# edit the date format on x-axis - Minor ticks
# Locating
ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0)) # set the day of week to Monday(0)
# Formatting
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d')) # this will show the date of Mondays only
plt.show()
```
