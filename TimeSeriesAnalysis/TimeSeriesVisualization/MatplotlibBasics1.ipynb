# Matplotlib - Visualization

```python
import matplotlib.pyplot as plt
```

```python
# show plots inline
%matplotlib inline
```

```python
# create some data to plot
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

# Functional Method

```python
plt.plot(x, y)
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Simple Plot Title')

plt.show()
```

```python
# multiple plots in the same canvas
plt.subplot(1, 2, 1)  # (rows, columns, panel number)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')

plt.show()
```

# Object Oriented API Method

Creat figure objects and call methods off of it

```python
fig = plt.figure() # create a figure object - this is an empty canvas

# add axes to the figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # left, bottom, width, height

#axes.plot(x, y)
axes.set_xlabel('X Axis Label')
axes.set_ylabel('Y Axis Label')
axes.set_title('Object Oriented Plot Title')

plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # left, bottom, width, height
axes.plot(x, y)
axes.set_xlabel('X Axis Label')
axes.set_ylabel('Y Axis Label')
axes.set_title('Object Oriented Plot Title')

plt.show()
```

## Sub-plots

```python
# creating multiple axes
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  # inset axes

plt.show()
```

```python
# creating multiple axes
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # main axes
axes2 = fig.add_axes([0.5, 0.15, 0.4, 0.3])  # inset axes

plt.show()
```

```python
# creating multiple axes
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  # inset axes

axes1.plot(x, y)
axes1.set_title('Larger Plot')

axes2.plot(y, x)
axes2.set_title('Smaller Plot')

plt.show()
```

# Subplots

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()

plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

# axes is an array of axes in an array so it can be iterated over
for current_ax in axes:
    current_ax.plot(x, y)
plt.tight_layout()

plt.show()
```

```python
# axes object
axes
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')    

plt.tight_layout()

plt.show()
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8, 2), dpi=100)  # width, height in inches
ax = fig.add_axes([0, 0, 1, 1])  # left, bottom, width, height (range 0 to 1)
ax.plot(x, y)

plt.show()
```

```python
# plotting subplots with specified figure size and DPI
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 4))

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
plt.show()
```

# Labeling the axises

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')
ax.set_title('Title')
plt.show()
```

# Legend

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='x squared')
ax.plot(x, x**3, label='x cubed')
#ax.legend(loc=0)  # best location
ax.legend(loc=(0.1, 0.1)) # custom location

plt.show()
```

# Plot Apperance

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='purple', lw=2, ls='--', marker='o', markersize=10, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')

# RGB hex color example
#ax.plot(x, y, color='#800080', lw=2, ls='--', marker='o', markersize=10, markerfacecolor='#FFFF00', markeredgewidth=3, markeredgecolor='#008000')

# alpha to control transparency
#ax.plot(x, y, color='purple', lw=2, ls='--', marker='o', markersize=10, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green', alpha=0.5)

# set x and y limits
ax.set_xlim([0, 1])
ax.set_ylim([0, 2])

plt.show()
```

# Special plots

```python
# scatter plot
plt.scatter(x, y)

plt.show()
```

```python
# histogram
from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data)
plt.show()
```

```python
# box plot - statistical distribution of categorial data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangle box plot    
plt.boxplot(data, vert=True, patch_artist=True)
plt.show()
```
